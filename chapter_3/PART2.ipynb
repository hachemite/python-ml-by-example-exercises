```python
import pandas as pd

n_rows = 300000
df = pd.read_csv("train", nrows=n_rows)
df.head()
```

```python
print(df.head(5))
```

```python
Y = df['click'].values
```

```python
X = df.drop(['click', 'id', 'hour', 'device_id', 'device_ip'],
axis=1).values
```

```python
print(X.shape)
```

```python
n_train = int(n_rows * 0.9)
X_train = X[:n_train]
Y_train = Y[:n_train]
X_test = X[n_train:]
Y_test = Y[n_train:]
```

one-hot encoding to convert categorical data (like website categories) into a numerical format that machine learning models can understand. Essentially, it turns each category into a separate binary feature (0 or 1).

the handle_unknown='ignore' parameter in the OneHotEncoder is used to prevent errors when the encoder encounters a category it hasn't seen before in the training data. If an unknown category is encountered, the resulting one-hot encoded features will all be set to 0.

```python
from sklearn.preprocessing import OneHotEncoder
enc = OneHotEncoder(handle_unknown='ignore')
```

```python
X_train_enc = enc.fit_transform(X_train)
X_train_enc[0]
```

```python
print(X_train_enc[0])
```

```python
X_test_enc = enc.transform(X_test)
```

(other hyperparam-
eters, such as min_samples_split and class_weight, are also highly recommended)

```python
from sklearn.tree import DecisionTreeClassifier
parameters = {'max_depth': [3, 10, None]} # Different depths to test
```

```python
decision_tree = DecisionTreeClassifier(criterion='gini', min_samples_split=30)
```

***Step 2: Perform GridSearchCV to find the best hyperparameters***

```python
from sklearn.model_selection import GridSearchCV

# Perform GridSearchCV to find the best hyperparameters
grid_search = GridSearchCV(
    decision_tree,          # Model to optimize
    parameters,             # Hyperparameters to search
    n_jobs=-1,              # Use all CPU cores for parallel computation
    cv=3,                   # 3-fold cross-validation
    scoring='roc_auc'       # Use AUC for model evaluation
)
```

Note, n_jobs=-1 means that we use all of the available CPU processors

```python
grid_search.fit(X_train_enc, Y_train)
print(grid_search.best_params_)
```

**Use the best model to make predictions on the test set**

```python
decision_tree_best = grid_search.best_estimator_  # Get the best model
pos_prob = decision_tree_best.predict_proba(X_test_enc)[:, 1]  # Predicted probabilities for class 1 (click)

from sklearn.metrics import roc_auc_score
print(f'The ROC AUC on testing set is: {roc_auc_score(Y_test,pos_prob):.3f}')
```

**Compare with a random predictor (baseline)**

A random predictor should give AUC ≈ 0.5 (no better than random guessing)


```python
import numpy as np
pos_prob = np.zeros(len(Y_test))
click_index = np.random.choice(len(Y_test),
int(len(Y_test) * 51211.0/300000), #Calculate the number of clicks to simulate (~17% click rate, based on dataset statistics
replace=False)

# Randomly select indices to assign as "click" (1)
pos_prob[click_index] = 1
print(f'The ROC AUC on testing set using random selection is: {roc_auc_score(Y_test, pos_prob):.3f}')
```

## **Ensembling decision trees – random forests**

**Initialize Random Forest with some default parameters:**
- n_estimators=100: creates 100 decision trees in the forest
- criterion='gini': uses Gini impurity to measure split quality
- min_samples_split=30: minimum samples required to split a node
- n_jobs=-1: uses all available CPU cores for parallel processing

```python
from sklearn.ensemble import RandomForestClassifier
random_forest = RandomForestClassifier(n_estimators=100,
                                       criterion='gini',
                                       min_samples_split=30,
                                       n_jobs=-1)
```

Initialize GridSearchCV:
- random_forest: the model to optimize
- parameters: hyperparameters to search
- n_jobs=-1: parallel processing
- cv=3: 3-fold cross-validation
- scoring='roc_auc': uses AUC as evaluation metric

WARNING: This may take a long time (20+ minutes) because:
1. Random Forest is computationally intensive (100 trees per configuration)
2. 5 max_depth values × 3 CV folds = 15 total fits
3. Large dataset size makes each fit slow

```python
parameters = {'max_depth': [5, 10, None]}  # Instead of many values



grid_search = GridSearchCV(random_forest,
                           parameters,
                           n_jobs=-1,
                           cv=3,
                           scoring='roc_auc')
grid_search.fit(X_train_enc, Y_train)
print(grid_search.best_params_)
```

```python
random_forest_best = grid_search.best_estimator_
pos_prob = random_forest_best.predict_proba(X_test)[:, 1]
print(f'The ROC AUC on testing set using random forest is: {roc_auc_score(Y_test, pos_prob):.3f}')
```

## 1.In the decision tree click-through prediction project, can you also tweak other hyperparameters,such as min_samples_split and class_weight? What is the highest AUC you are able to achieve?

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import GridSearchCV

# Expanded parameter grid
parameters = {
    'max_depth': [5, 10, 15, None],
    'min_samples_split': [2, 10, 30, 50],
    'class_weight': [None, 'balanced', {0:1, 1:3}],
    'min_samples_leaf': [1, 5, 10],
    'max_features': ['sqrt', 'log2']
}

# Initialize decision tree with gini impurity
decision_tree = DecisionTreeClassifier(criterion='gini', random_state=42)

# Set up grid search with ROC AUC scoring
grid_search = GridSearchCV(
    decision_tree,
    parameters,
    n_jobs=-1,
    cv=3,
    scoring='roc_auc',
    verbose=2  # Shows progress
)

# Fit the model
grid_search.fit(X_train_enc, Y_train)

# Show best parameters
print("Best parameters:", grid_search.best_params_)
```

**Evaluating the Best Model**

```python
# Get the optimized model
best_dt = grid_search.best_estimator_

# Make predictions
pos_prob = best_dt.predict_proba(X_test_enc)[:, 1]

# Calculate AUC
auc_score = roc_auc_score(Y_test, pos_prob)
print(f"Optimized Decision Tree AUC: {auc_score:.3f}")
```

## 2.In the random forest-based click-through prediction project, can you also tweak other hyper-parameters, such as min_samples_split, max_features, and n_estimators, in scikit-learn?

***What is the highest AUC you are able to achieve?***

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import RandomizedSearchCV  # Faster than GridSearchCV
import numpy as np

# Expanded parameter distribution
param_dist = {
    'n_estimators': [100, 200, 500],
    'max_depth': [5, 10, 15, None],
    'min_samples_split': [2, 5, 10, 30],
    'max_features': ['sqrt', 'log2', 0.3, 0.5],
    'class_weight': [None, 'balanced', {0:1, 1:2}]
}

# Initialize RF with warm_start=True for faster iterations
rf = RandomForestClassifier(
    criterion='gini',
    n_jobs=-1,
    random_state=42,
    warm_start=True
)

# Use RandomizedSearchCV for efficiency
random_search = RandomizedSearchCV(
    rf,
    param_distributions=param_dist,
    n_iter=30,  # Number of parameter combinations to try
    cv=3,
    scoring='roc_auc',
    n_jobs=-1,
    random_state=42,
    verbose=2
)

random_search.fit(X_train_enc, Y_train)

print("Best parameters:", random_search.best_params_)
```

**Evaluating the Best Model**

```python
best_rf = random_search.best_estimator_
pos_prob = best_rf.predict_proba(X_test_enc)[:, 1]
auc_score = roc_auc_score(Y_test, pos_prob)
print(f"Optimized Random Forest AUC: {auc_score:.3f}")
```
